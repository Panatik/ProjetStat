```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr
```

<h1 style = "color : #4682B4; font-weight:bold;">1- CH4</h1>

```python
ch4 = pd.read_csv("ch4_mm_gl.csv", skiprows=45)
```

<h3 style = "color : #4682B4; font-weight:bold;">1- Précisions sur le jeu de donnés du méthane CH4. </h3>

```python
print(f"Nombre total d'observations : {len(ch4)}")
```

```python
print(f"Période : de {ch4['year'].min()} à {ch4['year'].max()}")
```

<h3 style = "color : #4682B4; font-weight:bold;">2- Informations sur le méthane CH4. </h3>

Formule Chimique : CH4
Effet sur l'atmosphère : Le CH4 est un puissant gaz à effet de serre qui contribue fortement au réchauffement climatique.
Unité de mesure : PPM (Parties par million ex: 2ppm = 2 molécules de méthane pour 1 million de particules dans l'air ) 

<h3 style = "color : #4682B4; font-weight:bold;">3- Affichage du jeu de donnés du méthane CH4. </h3>

```python
ch4 = ch4.drop(columns=['year', 'decimal', 'average_unc', 'trend', 'trend_unc'])
ch4.head()
```

<h3 style = "color : #4682B4; font-weight:bold;">4- Graphe du méthane CH4. </h3>

```python
y = ch4['average']
x = ch4['month']

plt.figure(figsize=(10, 4))
plt.scatter(range(len(y)), y, s=10, color='green', alpha=0.6)
plt.title('Évolution du CH4 (ppm)')
plt.xlabel("Indice d'observation (mois)")
plt.ylabel("Concentration en CH4 (ppm)")
plt.grid(True)
plt.show()
```

<h3 style = "color : #4682B4; font-weight:bold;">5- 1) Moyenne mobile du méthane CH4. </h3>

```python
y_star = [np.nan] * len(y)

for t in range(6, len(y) - 6):
    window = y[t - 6 : t + 7].values
    if not np.isnan(window).any():
        moyenne = (
            0.5 * window[0] + 
            np.sum(window[1:12]) + 
            0.5 * window[12]
        ) / 12
        y_star[t] = moyenne

ch4['smoothed'] = y_star

plt.figure(figsize=(12, 5))
plt.plot(y, label="Série originale", alpha=0.6)
plt.plot(y_star, label="Moyenne mobile centrée (ordre 6)", color='green')
plt.xlabel("Indice des observations")
plt.ylabel("Concentration (ppb)")
plt.title("Méthane : Série lissée vs Originale")
plt.legend()
plt.grid(True)
plt.show()
```

Il n'est pas possible de la calculer pour les 6 premiers et 6 derniers mois car le calcul nécessite d'avoir des données pour les 6 mois précédants et suivants le mois calculé.
Cette moyenne permet d'estimer la tendance, sans les effets saisonniers.
C'est une "série lissée" car elle supprime le "bruit" (variations de court terme)

<h3 style = "color : #4682B4; font-weight:bold;">5- 2) Série privée du méthane CH4. </h3>

```python
ch4['private'] = ch4['average'] - ch4['smoothed']

plt.figure(figsize=(12, 4))
plt.plot(ch4['private'], label="Série privée", color='green')
plt.axhline(0, color='black', linestyle='--', alpha=0.5)
plt.xlabel("Indice d'observation")
plt.ylabel("Ecart")
plt.title("Série privée du méthane CH4")
plt.legend()
plt.grid(True)
plt.show()
```

Oscillations autour de 0 marquant les effets saisonniers avec des pics irréguliers (bruit).

<h3 style = "color : #4682B4; font-weight:bold;">5- 3) Coefficients mensuels du méthane CH4. </h3>

```python
data = ch4.dropna(subset=['private'])
monthly_coeffs = data.groupby('month')['private'].mean()

print(monthly_coeffs)
```

Cela représente la moyenne des fluctuations de chaques mois de chaques années. On peut voir qu'il y a un creux entre juin et août.

<h3 style = "color : #4682B4; font-weight:bold;">5- 4) Coefficients mensuels corrigés du méthane CH4. </h3>

```python
c_mean = monthly_coeffs.mean()
corrected_coeffs = monthly_coeffs - c_mean

print(corrected_coeffs)
```

Cela représente la variation saisonnière relative sans biais gloabal. On peut voir qu'il y a, par exemple, une surtendance en novembre à l'inverse de juillet.

<h3 style = "color : #4682B4; font-weight:bold;">5- 5) Série corrigée des variations saisonnières du méthane CH4. </h3>

```python
ch4['season_corr'] = ch4.apply(
    lambda row: row['average'] - corrected_coeffs.loc[row['month']] 
    if not np.isnan(row['average']) else np.nan,
    axis=1
)

plt.figure(figsize=(12, 5))
plt.plot(ch4['average'], label='Série initiale', alpha=0.5)
plt.plot(ch4['season_corr'], label='Série désaisonnalisée', color='green')
plt.title('Série corrigée des variations saisonnières')
plt.xlabel("Indice des observations (mois)")
plt.ylabel("Concentration du gaz")
plt.legend()
plt.grid(True)
plt.show()
```

Cette série ne contient plus les variations saisonnières et met en évidence la tendance/évolution (permet de mieux analyser des changements à long terme).

<h3 style = "color : #4682B4; font-weight:bold;">5- 6) Tendance du méthane CH4. </h3>

```python
# Création de l'indice temporel t (mois)
ch4 = ch4.dropna(subset=['season_corr'])  # on ne garde que les lignes valides
ch4['t'] = np.arange(len(ch4))  # indice 0, 1, ..., n-1

# Données pour la régression
X = ch4[['t']]  # variable explicative
y = ch4['season_corr']  # variable dépendante (série corrigée)

# Ajustement du modèle de régression linéaire
lr = LinearRegression()
lr.fit(X, y)

# Coefficients de la droite
a = lr.coef_[0]   # pente
b = lr.intercept_  # ordonnée à l'origine

# Valeurs prédites (tendance)
ch4['trend'] = lr.predict(X)

print(f"Équation de la tendance : y = {a}*t + {b}")
```

Signification: La concentration augmente de 0,112 ppm par mois (c'est le pas) et elle était de 1662,6696 ppm au départ.

```python
plt.figure(figsize=(12, 5))
plt.plot(ch4['season_corr'], label='Série corrigée', alpha=0.6)
plt.plot(ch4['trend'], label='Tendance', color='green')
plt.title("Tendance linéaire de la série série corrigée des variations saisonnières")
plt.xlabel("Mois")
plt.ylabel("Concentration du gaz")
plt.legend()
plt.grid(True)
plt.show()
```

Cela représente l'évolution moyenne à long terme du méthane CH4. Cette tendance ne tient pas en compte les fluctuations. Ainsi, lorsque la série est supérieure à la tendance cela veut dire que le gaz augmente en moyenne (et inversement).

<h3 style = "color : #4682B4; font-weight:bold;">5- 7) Prévision du méthane CH4 sur 2 ans. </h3>

```python
# Dernier indice temporel existant
last_t = ch4['t'].iloc[-1]

# Générer les 24 mois suivants
future_t = np.arange(last_t + 1, last_t + 25)
future_months = [(m % 12) + 1 for m in range(ch4['month'].iloc[-1], ch4['month'].iloc[-1] + 24)]

# Calcul de la tendance pour chaque t futur
future_trend = lr.predict(future_t.reshape(-1, 1))

# Appliquer les coefficients saisonniers corrigés
future_seasonality = [corrected_coeffs.loc[month] for month in future_months]

# Prévisions : tendance + saisonnalité
future_forecast = future_trend + future_seasonality

# Organiser en DataFrame
forecast_ch4 = pd.DataFrame({
    't': future_t,
    'month': future_months,
    'trend': future_trend,
    'seasonality': future_seasonality,
    'forecast': future_forecast
})

# Affichage
print(forecast_ch4[['month', 'forecast']])
```

```python
plt.figure(figsize=(12, 5))
plt.plot(ch4['t'], ch4['average'], label='Série originale', alpha=0.4)
plt.plot(ch4['t'], ch4['season_corr'], label='Série corrigée', alpha=0.6)
plt.plot(ch4['t'], ch4['trend'], label='Tendance', color='red')
plt.plot(forecast_ch4['t'], forecast_ch4['forecast'], label='Prévision (24 mois)', color='green')
plt.title("Prévision sur 2 ans")
plt.xlabel("Indice temporel t")
plt.ylabel("Concentration du gaz")
plt.legend()
plt.grid(True)
plt.show()
```

Cette méthode prend donc en compte la tendance et le comportement mensuel et ignore les fluctuations/changements.

<h3 style = "color : #4682B4; font-weight:bold;">5- 8) Observation du méthane CH4. </h3>

Elles peuvent être dues au cycle des plantes (ex: photosynthèse), aux variations de températures, à l'activité biologique ou encore aux comportements humains.

<h3 style = "color : #4682B4; font-weight:bold;">6- 1) Série lissée du méthane CH4. </h3>

```python
y_star = [np.nan] * len(y)

for t in range(6, len(y) - 6):
    window = y[t - 6 : t + 7].values
    if not np.isnan(window).any():
        moyenne = (
            0.5 * window[0] + 
            np.sum(window[1:12]) + 
            0.5 * window[12]
        ) / 12
        y_star[t] = moyenne

ch4['smoothed'] = y_star

plt.figure(figsize=(12, 5))
plt.plot(y, label="Série originale", alpha=0.6)
plt.plot(y_star, label="Moyenne mobile centrée (ordre 6)", color='green')
plt.xlabel("Indice des observations")
plt.ylabel("Concentration (ppb)")
plt.title("Méthane : Série lissée vs Originale")
plt.legend()
plt.grid(True)
plt.show()
```

<h3 style = "color : #4682B4; font-weight:bold;">6- 2) Série divisée du méthane CH4. </h3>

```python
ch4['divise'] = ch4['average']/ch4['smoothed']

plt.figure(figsize=(12, 4))
plt.plot(ch4['divise'], label="Série divise", color='green')
plt.axhline(0, color='black', linestyle='--', alpha=0.5)
plt.xlabel("Indice d'observation")
plt.ylabel("Ecart")
plt.title("Série divisée du méthane CH4")
plt.legend()
plt.grid(True)
plt.show()
```

Ici on évalue l’effet en proportion. C’est utile quand la série évolue. Si < 1: concentration en-dessous de la tendance. Si > 1: concentration au-dessus de la tendance.

<h3 style = "color : #4682B4; font-weight:bold;">6- 3) Coefficients mensuels du méthane CH4. </h3>

```python
data = ch4.dropna(subset=['divise'])
monthly_coeffs = data.groupby('month')['divise'].mean()

print(monthly_coeffs)
```

<h3 style = "color : #4682B4; font-weight:bold;">6- 4) Coefficients mensuels corrigés du méthane CH4. </h3>

```python
c_mean = monthly_coeffs.mean()
corrected_coeffs = monthly_coeffs/c_mean

print(corrected_coeffs)
```

Cela garantit que sur une année complète, l’effet saisonnier ne modifie pas la moyenne globale de la série. Cela reflète alors l’écart proportionnel au comportement moyen du gaz pour le mois j.

<h3 style = "color : #4682B4; font-weight:bold;">6- 5) Série corrigée des variations saisonnières du méthane CH4. </h3>

```python
ch4['season_corr'] = ch4.apply(
    lambda row: row['average']/corrected_coeffs.loc[row['month']] 
    if not np.isnan(row['average']) else np.nan,
    axis=1
)

plt.figure(figsize=(12, 5))
plt.plot(ch4['average'], label='Série initiale', alpha=0.5)
plt.plot(ch4['season_corr'], label='Série désaisonnalisée', color='green')
plt.title('Série corrigée des variations saisonnières')
plt.xlabel("Indice des observations (mois)")
plt.ylabel("Concentration du gaz")
plt.legend()
plt.grid(True)
plt.show()
```

Nous pouvons remarquer que nous avons le même résultat que pour la question 5. Cette série n'est n’est plus influencée par les effets saisonniers et supprime les oscillations (bruit).

<h3 style = "color : #4682B4; font-weight:bold;">6- 6) Tendance du méthane CH4. </h3>

```python
ch4 = ch4.dropna(subset=['season_corr'])
ch4['t'] = np.arange(len(ch4))

X = ch4[['t']]
y = ch4['season_corr']

lr = LinearRegression()
lr.fit(X, y)

a = lr.coef_[0]
b = lr.intercept_

ch4['trend'] = lr.predict(X)

print(f"Équation de la tendance : y = {a}*t + {b}")
```

```python
plt.figure(figsize=(12, 5))
plt.plot(ch4['season_corr'], label='Série corrigée', alpha=0.6)
plt.plot(ch4['trend'], label='Tendance', color='green')
plt.title("Tendance linéaire de la série série corrigée des variations saisonnières")
plt.xlabel("Mois")
plt.ylabel("Concentration du gaz")
plt.legend()
plt.grid(True)
plt.show()
```

<h3 style = "color : #4682B4; font-weight:bold;">6- 7) Prévision du méthane CH4 sur 2 ans. </h3>

```python
last_t = ch4['t'].iloc[-1]

future_t = np.arange(last_t + 1, last_t + 25)
future_months = [(m % 12) + 1 for m in range(ch4['month'].iloc[-1], ch4['month'].iloc[-1] + 24)]

future_trend = lr.predict(future_t.reshape(-1, 1))

future_seasonality = [corrected_coeffs.loc[month] for month in future_months]

future_forecast = future_trend * future_seasonality

forecast_ch4 = pd.DataFrame({
    't': future_t,
    'month': future_months,
    'trend': future_trend,
    'seasonality': future_seasonality,
    'forecast': future_forecast
})

print(forecast_ch4[['month', 'forecast']])
```

```python
plt.figure(figsize=(12, 5))
plt.plot(ch4['t'], ch4['average'], label='Série originale', alpha=0.4)
plt.plot(ch4['t'], ch4['season_corr'], label='Série corrigée', alpha=0.6)
plt.plot(ch4['t'], ch4['trend'], label='Tendance', color='red')
plt.plot(forecast_ch4['t'], forecast_ch4['forecast'], label='Prévision (24 mois)', color='green')
plt.title("Prévision sur 2 ans")
plt.xlabel("Indice temporel t")
plt.ylabel("Concentration du gaz")
plt.legend()
plt.grid(True)
plt.show()
```

La méthode est différente mais les résultats sont presque les même que précédemment. Ici on suppose que la saisonnalité agit proportionnellement à la tendance. Par exemple, si la concentration de CH4 est en augmentation générale, alors l'effet saisonnier aura un impact plus grand en valeur absolue qu'auparavant.

<h3 style = "color : #4682B4; font-weight:bold;">9- 6) Régression exponentielle du méthane CH4 sur 2 ans. </h3>

```python
X = yearly_ch4["t"].values.reshape(-1, 1)
y = yearly_ch4["average"].values

log_y = np.log(y)

exp_lr = LinearRegression()
exp_lr.fit(X, log_y)

a = exp_lr.coef_[0]

log_b = exp_lr.intercept_
b = np.exp(log_b)

print(f"Équation exponentielle : y = {b} * exp({a} * t)")
```

```python
y_exp_pred = b * np.exp(a * yearly_ch4["t"])

plt.figure(figsize=(10, 4))
plt.scatter(yearly_ch4["year"], yearly_ch4["average"], label="Données")
plt.plot(yearly_ch4["year"], y_exp_pred, color="green", label="Régression exponentielle")
plt.title("Régression exponentielle sur données annuelles")
plt.legend()
plt.show()
```

```python
future_pred_exp = b * np.exp(a * future_t.flatten())

forecast_exp_ch4 = pd.DataFrame({
    "t": future_t.flatten(),
    "forecast_exponential": future_pred_exp
})
print(forecast_exp_ch4)
```

Non car la concentration augmente à un rythme constant, le modèle linéaire est donc suffisant.

```python
# Création de l'indice temporel t (mois)
co2 = co2.dropna(subset=['season_corr'])  # on ne garde que les lignes valides
co2['t'] = np.arange(len(co2))  # indice 0, 1, ..., n-1

# Données pour la régression
X = co2[['t']]  # variable explicative
y = co2['season_corr']  # variable dépendante (série corrigée)

# Ajustement du modèle de régression linéaire
lr = LinearRegression()
lr.fit(X, y)

# Coefficients de la droite
a = lr.coef_[0]   # pente
b = lr.intercept_  # ordonnée à l'origine

# Valeurs prédites (tendance)
co2['trend'] = lr.predict(X)

# Affichage de l'équation
print(f"Équation de la tendance : y = {a}*t + {b}")
```

La concentration de CO₂ augmente en moyenne de 0,1375 ppm par mois. Cela reflète une hausse régulière et continue du CO₂ dans l’atmosphère, en plus des variations saisonnières.

```python
plt.figure(figsize=(12, 5))
plt.plot(co2['season_corr'], label='Série corrigée', alpha=0.6)
plt.plot(co2['trend'], label='Tendance', color='green')
plt.title("Tendance linéaire de la série série corrigée des variations saisonnières")
plt.xlabel("Mois")
plt.ylabel("Concentration du gaz")
plt.legend()
plt.grid(True)
plt.show()
```

Cela représente l'évolution moyenne à long terme du dioxyde de carbone CO2. Cette tendance ne tient pas en compte les fluctuations. Ainsi, lorsque la série est supérieure à la tendance cela veut dire que le gaz augmente en moyenne (et inversement).

<h3 style = "color : #4682B4; font-weight:bold;">5- 7) Prévision du dioxyde de carbone CO2 sur 2 ans. </h3>

```python
# Dernier indice temporel existant
last_t = co2['t'].iloc[-1]

# Générer les 24 mois suivants
future_t = np.arange(last_t + 1, last_t + 25)
future_months = [(m % 12) + 1 for m in range(co2['month'].iloc[-1], co2['month'].iloc[-1] + 24)]

# Calcul de la tendance pour chaque t futur
future_trend = lr.predict(future_t.reshape(-1, 1))

# Appliquer les coefficients saisonniers corrigés
future_seasonality = [corrected_coeffs.loc[month] for month in future_months]

# Prévisions : tendance + saisonnalité
future_forecast = future_trend + future_seasonality

# Organiser en DataFrame
forecast_co2 = pd.DataFrame({
    't': future_t,
    'month': future_months,
    'trend': future_trend,
    'seasonality': future_seasonality,
    'forecast': future_forecast
})

# Affichage
print(forecast_co2[['month', 'forecast']])
```

```python
plt.figure(figsize=(12, 5))
plt.plot(co2['t'], co2['average'], label='Série originale', alpha=0.4)
plt.plot(co2['t'], co2['season_corr'], label='Série corrigée', alpha=0.6)
plt.plot(co2['t'], co2['trend'], label='Tendance', color='red')
plt.plot(forecast_co2['t'], forecast_co2['forecast'], label='Prévision (24 mois)', color='green')
plt.title("Prévision sur 2 ans")
plt.xlabel("Indice temporel t")
plt.ylabel("Concentration du gaz")
plt.legend()
plt.grid(True)
plt.show()
```

Cette méthode prend donc en compte la tendance et le comportement mensuel et ignore les fluctuations/changements.

<h3 style = "color : #4682B4; font-weight:bold;">5- 8) Observation du dioxyde de carbone CO2. </h3>

Elles peuvent être dues au cycle des plantes (ex: photosynthèse), aux variations de températures, à l'activité biologique ou encore aux comportements humains et aux variations climatiques.

<h3 style = "color : #4682B4; font-weight:bold;">6- 1) Série lissée du dioxyde de carbone CO2. </h3>

```python
y_star = [np.nan] * len(y)

for t in range(6, len(y) - 6):
    window = y[t - 6 : t + 7].values
    if not np.isnan(window).any():
        moyenne = (
            0.5 * window[0] + 
            np.sum(window[1:12]) + 
            0.5 * window[12]
        ) / 12
        y_star[t] = moyenne

co2['smoothed'] = y_star

plt.figure(figsize=(12, 5))
plt.plot(y, label="Série originale", alpha=0.6)
plt.plot(y_star, label="Moyenne mobile centrée (ordre 6)", color='green')
plt.xlabel("Indice des observations")
plt.ylabel("Concentration (ppm)")
plt.title("dioxyde de carbone : Série lissée vs Originale")
plt.legend()
plt.grid(True)
plt.show()
```

<h3 style = "color : #4682B4; font-weight:bold;">6- 2) Série divisée du dioxyde de carbone CO2. </h3>

```python
co2['divise'] = co2['average']/co2['smoothed']

plt.figure(figsize=(12, 4))
plt.plot(co2['divise'], label="Série divise", color='green')
plt.axhline(0, color='black', linestyle='--', alpha=0.5)
plt.xlabel("Indice d'observation")
plt.ylabel("Ecart")
plt.title("Série divisée du dioxyde de carbone CO2")
plt.legend()
plt.grid(True)
plt.show()
```

Ici on évalue l’effet en proportion. C’est utile quand la série évolue. Si < 1: concentration en-dessous de la tendance. Si > 1: concentration au-dessus de la tendance.

<h3 style = "color : #4682B4; font-weight:bold;">6- 3) Coefficients mensuels du dioxyde de carbone CO2. </h3>

```python
data = co2.dropna(subset=['divise'])
monthly_coeffs = data.groupby('month')['divise'].mean()

print(monthly_coeffs)
```

<h3 style = "color : #4682B4; font-weight:bold;">6- 4) Coefficients mensuels corrigés du dioxyde de carbone CO2. </h3>

```python
c_mean = monthly_coeffs.mean()
corrected_coeffs = monthly_coeffs/c_mean

print(corrected_coeffs)
```

Ces coefficients montrent que le co2 est légèrement au-dessus de la tendance au printemps (jusqu’à +0,8 %) et légèrement en dessous en fin d’été (environ −1 %), traduisant une saisonnalité relative modérée.

<h3 style = "color : #4682B4; font-weight:bold;">6- 5) Série corrigée des variations saisonnières du dioxyde de carbone CO2. </h3>

```python
co2['season_corr'] = co2.apply(
    lambda row: row['average']/corrected_coeffs.loc[row['month']] 
    if not np.isnan(row['average']) else np.nan,
    axis=1
)

plt.figure(figsize=(12, 5))
plt.plot(co2['average'], label='Série initiale', alpha=0.5)
plt.plot(co2['season_corr'], label='Série désaisonnalisée', color='green')
plt.title('Série corrigée des variations saisonnières')
plt.xlabel("Indice des observations (mois)")
plt.ylabel("Concentration du gaz")
plt.legend()
plt.grid(True)
plt.show()
```

Nous pouvons remarquer que nous avons le même résultat que pour la question 5. Cette série n'est n’est plus influencée par les effets saisonniers et supprime les oscillations (bruit).

<h3 style = "color : #4682B4; font-weight:bold;">6- 6) Tendance du dioxyde de carbone CO2. </h3>

```python
co2 = co2.dropna(subset=['season_corr'])
co2['t'] = np.arange(len(co2))

X = co2[['t']]
y = co2['season_corr']

lr = LinearRegression()
lr.fit(X, y)

a = lr.coef_[0]
b = lr.intercept_

co2['trend'] = lr.predict(X)

print(f"Équation de la tendance : y = {a}*t + {b}")
```

```python
plt.figure(figsize=(12, 5))
plt.plot(co2['season_corr'], label='Série corrigée', alpha=0.6)
plt.plot(co2['trend'], label='Tendance', color='green')
plt.title("Tendance linéaire de la série série corrigée des variations saisonnières")
plt.xlabel("Mois")
plt.ylabel("Concentration du gaz")
plt.legend()
plt.grid(True)
plt.show()
```

<h3 style = "color : #4682B4; font-weight:bold;">6- 7) Prévision du dioxyde de carbone CO2 sur 2 ans. </h3>

```python
last_t = co2['t'].iloc[-1]

future_t = np.arange(last_t + 1, last_t + 25)
future_months = [(m % 12) + 1 for m in range(co2['month'].iloc[-1], co2['month'].iloc[-1] + 24)]

future_trend = lr.predict(future_t.reshape(-1, 1))

future_seasonality = [corrected_coeffs.loc[month] for month in future_months]

future_forecast = future_trend * future_seasonality

forecast_co2 = pd.DataFrame({
    't': future_t,
    'month': future_months,
    'trend': future_trend,
    'seasonality': future_seasonality,
    'forecast': future_forecast
})

print(forecast_co2[['month', 'forecast']])
```

```python
plt.figure(figsize=(12, 5))
plt.plot(co2['t'], co2['average'], label='Série originale', alpha=0.4)
plt.plot(co2['t'], co2['season_corr'], label='Série corrigée', alpha=0.6)
plt.plot(co2['t'], co2['trend'], label='Tendance', color='red')
plt.plot(forecast_co2['t'], forecast_co2['forecast'], label='Prévision (24 mois)', color='green')
plt.title("Prévision sur 2 ans")
plt.xlabel("Indice temporel t")
plt.ylabel("Concentration du gaz")
plt.legend()
plt.grid(True)
plt.show()
```

La méthode est différente, mais les résultats restent très proches de la précédente. Ici, on suppose que la saisonnalité agit de manière proportionnelle à la tendance : ainsi, si la concentration de co2 augmente globalement, l’effet saisonnier se traduit par des variations plus importantes en valeur absolue qu’avec la méthode additive.
