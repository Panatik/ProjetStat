<h3 style = "color : #4682B4; font-weight:bold;">5- 1) Moyenne mobile du méthane CH4. </h3>

```python
y_star = [np.nan] * len(y)

for t in range(6, len(y) - 6):
    window = y[t - 6 : t + 7].values
    if not np.isnan(window).any():
        moyenne = (
            0.5 * window[0] + 
            np.sum(window[1:12]) + 
            0.5 * window[12]
        ) / 12
        y_star[t] = moyenne

ch4['smoothed'] = y_star

plt.figure(figsize=(12, 5))
plt.plot(y, label="Série originale", alpha=0.6)
plt.plot(y_star, label="Moyenne mobile centrée (ordre 6)", color='green')
plt.xlabel("Indice des observations")
plt.ylabel("Concentration (ppb)")
plt.title("Méthane : Série lissée vs Originale")
plt.legend()
plt.grid(True)
plt.show()
```

<h3 style = "color : #4682B4; font-weight:bold;">5- 2) Série privée du méthane CH4. </h3>

```python
ch4['private'] = ch4['average'] - ch4['smoothed']

plt.figure(figsize=(12, 4))
plt.plot(ch4['private'], label="Série privée", color='green')
plt.axhline(0, color='black', linestyle='--', alpha=0.5)
plt.xlabel("Indice d'observation")
plt.ylabel("Ecart")
plt.title("Série privée du méthane CH4")
plt.legend()
plt.grid(True)
plt.show()
```

<h3 style = "color : #4682B4; font-weight:bold;">5- 6) Tendance du méthane CH4. </h3>

```python
# Création de l'indice temporel t (mois)
ch4 = ch4.dropna(subset=['season_corr'])  # on ne garde que les lignes valides
ch4['t'] = np.arange(len(ch4))  # indice 0, 1, ..., n-1

# Données pour la régression
X = ch4[['t']]  # variable explicative
y = ch4['season_corr']  # variable dépendante (série corrigée)

# Ajustement du modèle de régression linéaire
lr = LinearRegression()
lr.fit(X, y)

# Coefficients de la droite
a = lr.coef_[0]   # pente
b = lr.intercept_  # ordonnée à l'origine

# Valeurs prédites (tendance)
ch4['trend'] = lr.predict(X)

# Affichage de l'équation
print(f"Équation de la tendance : y = {a:.4f}*t + {b:.4f}")
```
