```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

<h1 style = "color : #4682B4; font-weight:bold;">1- CH4</h1>

```python
ch4 = pd.read_csv("ch4_mm_gl.csv", skiprows=45)
```

<h3 style = "color : #4682B4; font-weight:bold;">1- Précisions sur le jeu de donnés du méthane CH4. </h3>

```python
print(f"Nombre total d'observations : {len(ch4)}")
```

```python
print(f"Période : de {ch4['year'].min()} à {ch4['year'].max()}")
```

<h3 style = "color : #4682B4; font-weight:bold;">2- Informations sur le méthane CH4. </h3>

Formule Chimique : CH4
Effet sur l'atmosphère : Le CH4 est un puissant gaz à effet de serre qui contribue fortement au réchauffement climatique.
Unité de mesure : PPM (Parties par million ex: 2ppm = 2 molécules de méthane pour 1 million de particules dans l'air ) 

<h3 style = "color : #4682B4; font-weight:bold;">3- Affichage du jeu de donnés du méthane CH4. </h3>

```python
ch4 = ch4.drop(columns=['year', 'decimal', 'average_unc', 'trend', 'trend_unc'])
ch4.head()
```

<h3 style = "color : #4682B4; font-weight:bold;">4- Graphe du méthane CH4. </h3>

```python
y = ch4['average']
x = ch4['month']

plt.figure(figsize=(10, 4))
plt.scatter(range(len(y)), y, s=10, color='green', alpha=0.6)
plt.title('Évolution du CH4 (ppm)')
plt.xlabel("Indice d'observation (mois)")
plt.ylabel("Concentration en CH4 (ppm)")
plt.grid(True)
plt.show()
```

<h3 style = "color : #4682B4; font-weight:bold;">5- 1) Moyenne mobile du méthane CH4. </h3>

```python
y_star = [np.nan] * len(y)

for t in range(6, len(y) - 6):
    window = y[t - 6 : t + 7].values
    if not np.isnan(window).any():
        moyenne = (
            0.5 * window[0] + 
            np.sum(window[1:12]) + 
            0.5 * window[12]
        ) / 12
        y_star[t] = moyenne

ch4['smoothed'] = y_star

plt.figure(figsize=(12, 5))
plt.plot(y, label="Série originale", alpha=0.6)
plt.plot(y_star, label="Moyenne mobile centrée (ordre 6)", color='green')
plt.xlabel("Indice des observations")
plt.ylabel("Concentration (ppb)")
plt.title("Méthane : Série lissée vs Originale")
plt.legend()
plt.grid(True)
plt.show()
```

<h3 style = "color : #4682B4; font-weight:bold;">5- 2) Série privée du méthane CH4. </h3>

```python
ch4['private'] = ch4['average'] - ch4['smoothed']

plt.figure(figsize=(12, 4))
plt.plot(ch4['private'], label="Série privée", color='green')
plt.axhline(0, color='black', linestyle='--', alpha=0.5)
plt.xlabel("Indice d'observation")
plt.ylabel("Ecart")
plt.title("Série privée du méthane CH4")
plt.legend()
plt.grid(True)
plt.show()
```

<h3 style = "color : #4682B4; font-weight:bold;">5- 3) Coefficients mensuels du méthane CH4. </h3>

```python
data = ch4.dropna(subset=['private'])
monthly_coeffs = data.groupby('month')['private'].mean()

print(monthly_coeffs)
```

<h3 style = "color : #4682B4; font-weight:bold;">5- 4) Coefficients mensuels corrigés du méthane CH4. </h3>

```python
c_mean = monthly_coeffs.mean()
corrected_coeffs = monthly_coeffs - c_mean

print(corrected_coeffs)
```

<h3 style = "color : #4682B4; font-weight:bold;">5- 5) Série corrigée des variations saisonnières du méthane CH4. </h3>

```python
ch4['season_corr'] = ch4.apply(
    lambda row: row['average'] - corrected_coeffs.loc[row['month']] 
    if not np.isnan(row['average']) else np.nan,
    axis=1
)

plt.figure(figsize=(12, 5))
plt.plot(ch4['average'], label='Série initiale', alpha=0.5)
plt.plot(ch4['season_corr'], label='Série désaisonnalisée', color='green')
plt.title('Série corrigée des variations saisonnières')
plt.xlabel("Indice des observations (mois)")
plt.ylabel("Concentration du gaz")
plt.legend()
plt.grid(True)
plt.show()
```

<h3 style = "color : #4682B4; font-weight:bold;">5- 6) Tendance du méthane CH4. </h3>

```python
# Création de l'indice temporel t (mois)
ch4 = ch4.dropna(subset=['season_corr'])  # on ne garde que les lignes valides
ch4['t'] = np.arange(len(ch4))  # indice 0, 1, ..., n-1

# Données pour la régression
X = ch4[['t']]  # variable explicative
y = ch4['season_corr']  # variable dépendante (série corrigée)

# Ajustement du modèle de régression linéaire
lr = LinearRegression()
lr.fit(X, y)

# Coefficients de la droite
a = lr.coef_[0]   # pente
b = lr.intercept_  # ordonnée à l'origine

# Valeurs prédites (tendance)
ch4['trend'] = lr.predict(X)

# Affichage de l'équation
print(f"Équation de la tendance : y = {a:.4f}*t + {b:.4f}")
```

```python
plt.figure(figsize=(12, 5))
plt.plot(ch4['season_corr'], label='Série corrigée', alpha=0.6)
plt.plot(ch4['trend'], label='Tendance', color='green')
plt.title("Tendance linéaire de la série série corrigée des variations saisonnières")
plt.xlabel("Mois")
plt.ylabel("Concentration du gaz")
plt.legend()
plt.grid(True)
plt.show()
```

<h3 style = "color : #4682B4; font-weight:bold;">5- 7) Prévision du méthane CH4 sur 2 ans. </h3>

```python
# Dernier indice temporel existant
last_t = ch4['t'].iloc[-1]

# Générer les 24 mois suivants
future_t = np.arange(last_t + 1, last_t + 25)
future_months = [(m % 12) + 1 for m in range(ch4['month'].iloc[-1], ch4['month'].iloc[-1] + 24)]

# Calcul de la tendance pour chaque t futur
future_trend = lr.predict(future_t.reshape(-1, 1))

# Appliquer les coefficients saisonniers corrigés
future_seasonality = [corrected_coeffs.loc[month] for month in future_months]

# Prévisions : tendance + saisonnalité
future_forecast = future_trend + future_seasonality

# Organiser en DataFrame
forecast_ch4 = pd.DataFrame({
    't': future_t,
    'month': future_months,
    'trend': future_trend,
    'seasonality': future_seasonality,
    'forecast': future_forecast
})

# Affichage
print(forecast_ch4[['month', 'forecast']])
```

```python
plt.figure(figsize=(12, 5))
plt.plot(ch4['t'], ch4['average'], label='Série originale', alpha=0.4)
plt.plot(ch4['t'], ch4['season_corr'], label='Série corrigée', alpha=0.6)
plt.plot(ch4['t'], ch4['trend'], label='Tendance', color='red')
plt.plot(forecast_ch4['t'], forecast_ch4['forecast'], label='Prévision (24 mois)', color='green')
plt.title("Prévision sur 2 ans")
plt.xlabel("Indice temporel t")
plt.ylabel("Concentration du gaz")
plt.legend()
plt.grid(True)
plt.show()
```
